## Cohort Analysis

```python
user_return = {}
with open('/home/hunor/jdsa_final_task/read_return.csv') as csvfile:
    readerCSV = csv.reader(csvfile, delimiter=';')
    for row in readerCSV:
       if row[4] not in user_return:
           user_return[row[4]] = []
       if row[0] not in user_return[row[4]]:
           user_return[row[4]].append(row[0])
```

```python
#user_return
```

```python
user_return_list = []
```

```python
for user in user_return:
    user_return_list.append(user_return[user])
```

```python
user_return_list
```

```python
start_date = datetime(2018, 1, 1)

date_list =[]

for i in range(90):
    current_date = start_date + timedelta(days=i)
    formatted_date = current_date.strftime('%Y-%m-%d')
    date_list.append(formatted_date)
```

```python
date_list
```

```python
results_df = pd.DataFrame(columns=['cohort', 'days_passed', 'number_of_users'])
```

```python
print("cohort;days_passed;number_of_users")
for cohort in date_list:
    day_passed = 0
    for given_date in date_list[date_list.index(cohort):date_list.index(cohort)+8]:
        number_of_users = 0
        for users in user_return_list:
            if given_date in users and users[0] == cohort:
                number_of_users = number_of_users + 1
        #results_df = results_df.append({'cohort': cohort, 'days_passed': day_passed, 'number_of_users': number_of_users}, ignore_index=True)
        print(cohort + ";" + str(day_passed) + ";" + str(number_of_users))
        day_passed = day_passed + 1

#results_df.to_csv('cohort_analysis.csv', index=False)
```

```python
first_reader_tmp = first_reader.groupby('my_date').count()[['user_id']]
first_reader_tmp = first_reader_tmp.reset_index()
first_reader_tmp
```

```python
x = first_reader_tmp.index
y = first_reader_tmp.user_id
```

```python
coefs_1 = np.polyfit(x, y, 1)
predict_1 = np.poly1d(coefs_1)
coefs_2 = np.polyfit(x, y, 2)
predict_2 = np.poly1d(coefs_2)
```

```python
x_test = np.linspace(0, 90)
y_pred = predict_1(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()
```
